# Interactive Tutorial: Building an Echo Bot

This tutorial will guide you through the process of building a simple "echo bot" using the `signal-client` library. An echo bot is a bot that responds to any message it receives by sending the same message back to the sender.

## 1. Installation

First, you'll need to install the `signal-client` library. You can do this using `pip`:

```python
!pip install signal-client
```

## 2. Configuration

Next, you'll need to configure the bot with your Signal account details. You can do this by creating a `config.yml` file with the following content:

```python
%%writefile config.yml
signal_service: "<your_signal_service>"
phone_number: "<your_phone_number>"
```

## 3. The Code

Now, let's write the code for the bot. We'll create a new file called `echo_bot.py` and add the following code:

```python
%%writefile echo_bot.py
from signal_client.bot import Bot

bot = Bot("echo_bot")

@bot.command("echo")
async def echo(context):
    await context.send(context.message.text)

if __name__ == "__main__":
    bot.run()
```

## 4. Running the Bot

Finally, you can run the bot by executing the following command in your terminal:

```python
!python echo_bot.py
```
